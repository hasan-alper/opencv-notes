# Drawing on Images

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt 
```

```python
blank_img = np.zeros(shape=(512, 512, 3), dtype=np.int16)
```

```python
plt.imshow(blank_img)
```

## Shapes

### Rectangle
* img: Image.
* pt1: Top left vertex of the rectangle.
* pt2: Bottom right vertex of the rectangle opposite to pt1.
* color: Rectangle color or brightness (grayscale image).
* thickness: Thickness of lines that make up the rectangle. -1 for filled rectangle.

```python
cv2.rectangle(blank_img, pt1=(200, 250), pt2=(350, 400), color=(0, 255, 0), thickness=6)
plt.imshow(blank_img)
```

### Circle
* img: Image.
* center: Center of the circle.
* radius: Radius of the circle
* color: Circle color or brightness (grayscale image).
* thickness: Thickness of the circle. -1 for filled circle.

```python
cv2.circle(img=blank_img, center=(100, 100), radius=50, color=(255, 0, 0), thickness=-1)
plt.imshow(blank_img)
```

### Line
* img: Image.
* pt1: Starting point of the line.
* pt2: Ending point of the line.
* color: Line color or brightness (grayscale image).
* thickness: Thickness of the line.

```python
cv2.line(img=blank_img, pt1=(0, 0), pt2=(511, 511), color=(0, 255, 255), thickness=2)
plt.imshow(blank_img)
```

### Text
* img: Image.
* text: Text string to be drawn.
* org: Bottom-left corner of the text string in the image.
* fontFace: Font type.
* fontScale: Font scale factor that is multiplied by the font-specific base size.
* color: Text color. 
* thickness: Thickness of the lines used to draw a text.
* lineType: Line type.

```python
cv2.putText(img=blank_img, text="hello", org=(10, 500), fontFace=cv2.FONT_HERSHEY_SIMPLEX, fontScale=3, color=(255 ,255, 255), thickness=2, lineType=cv2.LINE_AA)
plt.imshow(blank_img)
```

### Line

To draw a polygon, first you need coordinates of vertices. Make those points into an array of shape ROWSx1x2 where ROWS are number of vertices and it should be of type int32. 

* img: Image.
* pts: Array of polygonal curves.
* isClosed: Flag indicating whether the drawn polylines are closed or not. If they are closed, the function draws a line from the last vertex of each curve to its first vertex.
* color: Polyline color.
* param: thickness Thickness of the polyline edges.
* lineType: Type of the line segments.

```python
vertices = np.array([[250, 50], [500, 100], [400, 300], [350, 150]], np.int32)
points = vertices.reshape((-1, 1, 2))

cv2.polylines(img=blank_img, pts=[points], isClosed=True, color=(255, 0, 255), thickness=5)
plt.imshow(blank_img)
```
