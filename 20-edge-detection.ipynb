# Edge Detection

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
bgr_img = cv2.imread("DATA/sammy_face.jpg")
plt.imshow(bgr_img)
```

## Bluring Before Edge Detection

```python
blurred_img = cv2.blur(bgr_img, ksize=(5, 5))
```

## Choosing Thresholds

```python
med_val = np.median(bgr_img) 
lower = 0.7 * med_val
upper = 1.3 * med_val
```

## Apply Canny Edge Detection

```python
edges = cv2.Canny(image=blurred_img, threshold1=lower, threshold2=upper + 50) # Play with numbers
plt.imshow(edges)
```
