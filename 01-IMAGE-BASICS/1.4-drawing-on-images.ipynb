# Drawing on Images

We can draw different geometric shapes with OpenCV.

[Drawing Functions in OpenCV](https://docs.opencv.org/4.x/dc/da5/tutorial_py_drawing_functions.html)

## Importing

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt 
```

## Creating a Blank Image

Blank image is basically a matrix filled with zeros. Also since every color in an image is represented by one byte, we should set it's dtype to uint8.

```python
blank_img = np.zeros(shape=(512, 512, 3), dtype=np.uint8)
plt.imshow(blank_img)
```

## Drawing Rectangle

To draw a rectangle, you need top-left corner and bottom-right corner of rectangle.

```python
cv2.rectangle(blank_img, pt1=(200, 250), pt2=(350, 400), color=(0, 255, 0), thickness=6)
plt.imshow(blank_img)
```

## Drawing Circle

To draw a circle, you need its center coordinates and radius.

```python
cv2.circle(img=blank_img, center=(100, 100), radius=50, color=(255, 0, 0), thickness=-1)
plt.imshow(blank_img)
```

## Drawing Line

To draw a line, you need to pass starting and ending coordinates of line.

```python
cv2.line(img=blank_img, pt1=(0, 0), pt2=(511, 511), color=(0, 255, 255), thickness=2)
plt.imshow(blank_img)
```

## Adding Text

You can also put texts in images.

```python
cv2.putText(img=blank_img, text="hello", org=(10, 500), fontFace=cv2.FONT_HERSHEY_SIMPLEX, fontScale=3, color=(255 ,255, 255), thickness=2, lineType=cv2.LINE_AA)
plt.imshow(blank_img)
```

## Drawing Polygon

To draw a polygon, first you need coordinates of vertices. Make those points into an array of shape ROWSx1x2 where ROWS are number of vertices and it should be of type int32.

```python
vertices = np.array([[250, 50], [500, 100], [400, 300], [350, 150]], np.int32)
points = vertices.reshape((-1, 1, 2))

cv2.polylines(img=blank_img, pts=[points], isClosed=True, color=(255, 0, 255), thickness=5)
plt.imshow(blank_img)
```
