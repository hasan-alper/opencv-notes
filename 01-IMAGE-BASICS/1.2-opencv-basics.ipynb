# OpenCV Basics

OpenCV is a library of Python bindings designed to solve computer vision problems. 

## Importing

As a first step, the OpenCV python library is imported. The proper way to do this is to additionally assign it the name cv, which is used in the following to reference the library.

```python
import matplotlib.pyplot as plt
import cv2
```

## Loading an Image

We read the image "puppy.jpg" from the DATA folder. In order to do so, a call to the `cv2.imread` function loads the image using the file path specified by the first argument. The second argument is optional and specifies the format in which we want the image.

`plt.imshow` displays data as an image.

```python
img = cv2.imread("../DATA/puppy.jpg")
plt.imshow(img)
```

## Converting from BGR to RGB

The image has been correctly loaded by OpenCV as a NumPy array, but the color of each pixel has been sorted as BGR.
Matplotlib's plot expects an RGB image so, for a correct display of the image, it is necessary to swap those channels. 
This operation can be done either by using OpenCV conversion functions `cv2.cvtColor` or by working directly with the NumPy array.

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

## Converting to Grayscale

As first input, `cv2.cvtColor` receives the original image. As second input, it receives the color space conversion code. Since we want to convert our original image from the BGR color space to gray, we use the code `COLOR_BGR2GRAY`.

If you want to display grayscale using matplotlib, set `cmap` parameter to `"gray"`.

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(img_gray, cmap="gray")
```

## Resizing Images

Resizing an image means changing the dimensions of it, be it width alone, height alone or changing both of them. Also, the aspect ratio of the original image could be preserved in the resized image. To resize an image, OpenCV provides `cv2.resize` function.

### Resize to specific width and height

We will provide specific value in pixels for both width and height.

```python
img = cv2.resize(img_rgb, (1300, 500))
plt.imshow(img)
```

### Resize and preserve aspect ratio

We will provide percentage by which image has to be scaled.

```python
img = cv2.resize(img_rgb, (0,0), img_rgb, 0.6, 0.8) # Image, (0,0), Image, Width Ratio, Height Ratio
plt.imshow(img)
```

## Flipping images

OpenCV provides methods to flip an image across its x- or y-axis.

* Along central x-axis: `0`
* Along central y-axis: `1`
* Along both axis: `-1`

```python
img = cv2.flip(img_rgb, 0)
```

## Saving images

```python
cv2.imwrite("flipped-puppy.jpg", img) # BGR version of the image
```
