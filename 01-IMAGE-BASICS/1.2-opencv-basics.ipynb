# OpenCV Basics

OpenCV is a library of Python bindings designed to solve computer vision problems. 

## Importing

As a first step, the OpenCV python library is imported. The proper way to do this is to additionally assign it the name `cv`, which is used in the following to reference the library.

```python
import matplotlib.pyplot as plt
import cv2
```

## Loading an Image

We read the image "puppy.jpg" from the DATA folder. In order to do so, a call to the `cv2.imread` function loads the image using the file path specified by the first argument. The second argument is optional and specifies the format in which we want the image.

`plt.imshow` displays data as an image.

---

`cv2.imread(string filename, int flags=cv2.IMREAD_COLOR)`

**Description:** This method loads an image from the specified file. If the image cannot be read (because of missing file, improper permissions, unsupported or invalid format) then this method returns an empty matrix.

**Parameters:**

`filename`: A string representing the path of the image to be read.

`flags`: It specifies the way in which image should be read.

* `cv2.IMREAD_COLOR`: It specifies to load a color image. Any transparency of image will be neglected. Alternatively, we can pass integer value 1 for this flag.

* `cv2.IMREAD_GRAYSCALE`: It specifies to load an image in grayscale mode. Alternatively, we can pass integer value 0 for this flag.

* `cv2.IMREAD_UNCHANGED`: It specifies to load an image as such including alpha channel. Alternatively, we can pass integer value -1 for this flag.

* and some more...

**Return Value:** This method returns an image that is loaded from the specified file.

---

```python
img = cv2.imread("../DATA/puppy.jpg")
plt.imshow(img)
```

## Converting Colorspaces

The image has been correctly loaded by OpenCV as a NumPy array, but the color of each pixel has been sorted as BGR.
Matplotlib's plot expects an RGB image so, for a correct display of the image, it is necessary to swap those channels. 
This operation can be done either by using OpenCV conversion functions `cv2.cvtColor` or by working directly with the NumPy array.

### Convert from BGR to RGB

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

### Convert from BGR to Grayscale

As first input, `cv2.cvtColor` receives the original image. As second input, it receives the color space conversion code. Since we want to convert our original image from the BGR color space to gray, we use the code `COLOR_BGR2GRAY`.

If you want to display grayscale using matplotlib, set `cmap` parameter to `"gray"`.

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(img_gray, cmap="gray")
```

## Resizing Images

Resizing an image means changing the dimensions of it, be it width alone, height alone or changing both of them. Also, the aspect ratio of the original image could be preserved in the resized image. To resize an image, OpenCV provides `cv2.resize` function.

---

`cv2.resize(Mat src, Tuple[int, int] dsize, Mat dst=None, double fx=0, double fy=0, int interpolation=cv2.INTER_LINEAR)`

**Description:** Resizes an image.

**Parameters:**

`src`: Input image array (Single-channel, 8-bit or floating-point) 

`dsize`: Size of the output array

`dst`: Output array

`fx`: Scale factor along the horizontal axis

`fy`: Scale factor along the vertical axis

`interpolation`: One of the above interpolation methods

* `INTER_AREA`: This is used when we need to shrink an image.
* `INTER_CUBIC`: This is slow but more efficient.
* `INTER_LINEAR`: This is primarily used when zooming is required.
* and some more...

**Return Value:** It returns an image.

---

### Resize to specific width and height

We will provide specific value in pixels for both width and height.

```python
img = cv2.resize(img_rgb, (1300, 500))
plt.imshow(img)
```

### Resize and preserve aspect ratio

We will provide percentage by which image has to be scaled.

```python
img = cv2.resize(img_rgb, None, fx=0.6, fy=0.6)
plt.imshow(img)
```

## Flipping Images

OpenCV provides methods to flip an image across its x- or y-axis.

---

`cv2.flip(Mat src, int flipCode, Mat dst=None)`

**Description:** Flips an image.

**Parameters:**

`src`: Input array.

`flipCode`: A flag to specify how to flip the array; 0 means flipping around the x-axis and positive value (for example, 1) means flipping around y-axis. Negative value (for example, -1) means flipping around both axes.

`dst`: Output array of the same size and type as src.

**Return Value:** It returns an image.

---

```python
img = cv2.flip(img_rgb, 0)
plt.imshow(img)
```

## Rotating Images

`cv2.rotate` method is used to rotate a 2D array in multiples of 90 degrees.

---

`cv2.rotate(Mat src, int rotateCode, Mat dst=None)`

**Description:** Rotates an image.

**Parameters:**

`src`: Input array.

`rotateCode`: It is an enum to specify how to rotate the array.

* `cv2.ROTATE_90_CLOCKWISE`: Rotates the image in a clockwise direction by 90 degrees.
* `cv2.ROTATE_90_COUNTERCLOCKWISE`: Rotates the image in a counter-clockwise direction by 90 degrees.
* `cv2.ROTATE_180`: Rotates the image in a clockwise direction by 180 degrees.

`dst`: Output array of the same size and type as src.

**Return Value:** It returns an image.


```python
img = cv2.rotate(img_rgb, cv2.ROTATE_90_COUNTERCLOCKWISE)
plt.imshow(img)
```

## Saving Images

---

`cv2.imwrite(string filename, Mat img, List[int] params)`

**Description:** Saves an image to a specified file.

**Parameters:**

`filename`: A string representing the file name. The filename must include image format like .jpg, .png, etc.

`img`: It is the image that is to be saved.

`params`: Format-specific parameters encoded as pairs (paramId_1, paramValue_1, paramId_2, paramValue_2, ... .) 

**Return Value:** It returns true if image is saved successfully.

---

```python
cv2.imwrite("flipped-puppy.jpg", img) # BGR version of the image
```
