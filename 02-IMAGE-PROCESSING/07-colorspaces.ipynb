# Colorspaces

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread('../DATA/puppy.jpg')
```

```python
rgb_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_img)
```

```python
hsv_img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
plt.imshow(hsv_img)
```

```python
hsl_img = cv2.cvtColor(img, cv2.COLOR_BGR2HLS)
plt.imshow(hsl_img)
```
