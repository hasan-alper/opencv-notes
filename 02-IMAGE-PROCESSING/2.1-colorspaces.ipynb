# Colorspaces

A colorspace is a specific organization of colors. It is a useful conceptual tool for understanding the color capabilities of a particular digital file. We will learn how to convert images from one color-space to another using OpenCV.

## Importing

```python
import cv2
import matplotlib.pyplot as plt
```

## Loading the Image

```python
img = cv2.imread("../DATA/puppy.jpg")
plt.imshow(img)
```

## Changing Colorspace

There are more than 150 color-space conversion methods available in OpenCV. But we will look most widely used ones such as BGR ↔ Gray and BGR ↔ HSV.

For color conversion, we use the function `cv.cvtColor`.

---

`cv2.cvtColor(Mat src, int code, Mat dst=None, int dstCn=0)`

**Description:** This method is used to convert an image from one color space to another.

**Parameters:**

`src`: Input image: 8-bit unsigned, 16-bit unsigned ( CV_16UC... ), or single-precision floating-point.

`code`: Color space conversion code.

* `COLOR_BGR2RGB`: convert between RGB and BGR color spaces
* `COLOR_BGR2GRAY`: convert between RGB/BGR and grayscale
* and 150 something more...

`dst`: Output image of the same size and depth as src.

`dstCn`: Number of channels in the destination image; if the parameter is 0, the number of the channels is derived automatically from src and code.


**Return Value:** It returns an image.


---

### Convert BGR to RGB

```python
rgb_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_img)
```

### Convert BGR to Grayscale

```python
hsv_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(hsv_img, cmap="gray")
```

### Convert BGR to HSV

```python
hsl_img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
plt.imshow(hsl_img)
```
